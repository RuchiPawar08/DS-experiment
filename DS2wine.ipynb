```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
df = pd.read_csv('/content/winequality-red.csv', delimiter=';')
df.head()
```

```python
%matplotlib inline
plt.xlabel('alcohol')
plt.ylabel('pH')
plt.scatter(df['alcohol'],df['pH'],color='red',marker='*')

# Create and fit a linear regression model
reg = linear_model.LinearRegression()
reg.fit(df[['alcohol']], df['pH'])

plt.plot(df['alcohol'],reg.predict(df[['alcohol']]),color='blue')
```
