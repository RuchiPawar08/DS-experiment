```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
df = pd.read_csv('/content/hiring.csv')
df.head()
```

```python
%matplotlib inline
df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].median(), inplace=True)
df['experience'].fillna('zero', inplace=True)
reg_test_score = linear_model.LinearRegression()
reg_test_score.fit(df[['test_score(out of 10)']], df['salary($)'])
plt.xlabel('test_score(out of 10)')
plt.ylabel('salary($)')
plt.scatter(df['test_score(out of 10)'],df['salary($)'],color='red',marker='*')
plt.plot(df['test_score(out of 10)'], reg_test_score.predict(df[['test_score(out of 10)']]),color='blue')
```
